# Food Choices Dataset Analysis




## Introduction

The Food Choices dataset is a comprehensive collection of data related to individuals' food preferences, habits, and associated lifestyle factors. This dataset is useful for analyzing various aspects of food choices, including dietary patterns, caloric intake, exercise habits, and their relationship with demographic factors such as gender, income, and marital status.



## Project Description

This project aims to explore and analyze the Food Choices dataset to gain insights into the various factors influencing individuals' food preferences and habits. The analysis will cover multiple dimensions such as dietary patterns, caloric intake, exercise habits, and their correlation with demographic attributes like gender, income, and marital status. The ultimate goal is to develop predictive models that can forecast food choices based on these attributes.



## Objectives

1. **Understand Dietary Patterns**: Analyze how different demographic groups prefer different types of food.
    
    
2. **Caloric Intake Analysis**: Study the caloric intake of individuals and how it varies with lifestyle choices.
    
    
3. **Exercise and Food Choices**: Investigate the relationship between exercise habits and food preferences.
    
    
4. **Marital Status and Eating Habits**: Examine how marital status influences food choices and eating habits.
    
    
5. **Gender-based Analysis**: Perform a gender-based analysis of food choices to uncover any significant differences.
    
    
6. **Self-Perception and Eating Habits**: Explore how self-perception of weight impacts food choices and dietary habits.
    
    
7. **Predictive Modeling**: Develop predictive models to understand and predict food preferences based on various factors.

### Dropping the unwanted columns

#### We can not drop all the null values because we will loose most of the important data
#### That's why we are replacing that null values with the mode values of perticular columns

```python
Mode = ['comfort_food_reasons_coded','calories_scone','cook','cuisine','drink','employment','exercise','fav_cuisine',
        'fav_food','income','marital_status','persian_food','self_perception_weight','soup','sports','tortilla_calories',
        'weight','GPA','calories_day',]
```

```python
for i in df.columns:
    if i in Mode:
        df[i].fillna(df[i].mode()[0], inplace=True)
```

```python
df.isna().sum()
```

### Correcting the Names of Fav_Cus

### Conclusion

In this notebook, we have analyzed the impact of different food choices on weight variations. Through our exploration of various datasets and visualizations, we have derived the following key insights:

- **Caloric Intake:** There is a clear relationship between caloric intake from different food items and weight changes. Higher calorie foods tend to be associated with weight gain.


- **Nutrient Composition:** The composition of nutrients (proteins, fats, carbohydrates) in the food choices significantly affects weight variations. Foods with balanced nutrient profiles contribute to healthier weight management.


- **Diet Patterns:** Different dietary patterns (e.g., high-protein vs. high-carbohydrate diets) have varying impacts on weight. Personal dietary preferences and lifestyle choices play a crucial role in determining these effects.

Overall, our analysis highlights the importance of mindful food choices and balanced diets in maintaining a healthy weight.
