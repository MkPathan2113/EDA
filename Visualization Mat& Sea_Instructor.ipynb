In matplotlib, the **figsize** attribute allows you to specify the width and height of a figure in inches. It's a tuple of the form **(width, height)**.

The default size is usually **[6.4, 4.8]** (6.4 inches in width and 4.8 inches in height). By adjusting figsize, you can ensure your plot has the desired size, especially when saving it to a file or displaying in a specific context.

````python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Default figsize
plt.figure()
plt.plot(x, y)
plt.title("Default figsize")
plt.show()

# Larger figsize
plt.figure(figsize=(10, 5))  # 10 inches in width, 5 inches in height
plt.plot(x, y)
plt.title("Larger figsize (10,5)")
plt.show()

# Smaller figsize
plt.figure(figsize=(4, 2))  # 4 inches in width, 2 inches in height
plt.plot(x, y)
plt.title("Smaller figsize (4,2)")
plt.show()
````

## **Box Plot**

**ðŸ“Š What is a box plot?**

1. A box plot, also known as a box-and-whisker plot, is a graphical representation of numerical data through their quartiles.

2. It shows the median, the upper and lower quartiles, and the outliers of a dataset.

3. The box plot provides a compact and clear summary of the distribution of a dataset and can be used to identify potential outliers.

**ðŸ“ˆ How to create a box plot using Python and Matplotlib?**

We can use the Matplotlib library in Python to create box plots.

Here is a code example that demonstrates how to create a simple box plot using Matplotlib:

````python
import matplotlib.pyplot as plt
import numpy as np

# Generate some random data
data = np.random.normal(size=(100,))

# Create a box plot
fig, ax = plt.subplots()
ax.boxplot(data)
ax.set_title('Box plot')
plt.show()
````

In this code example, we first generate some random data using NumPy. Then, we create a box plot using the boxplot() function of Matplotlib. Finally, we set a title for the plot and display it using the show() function of Matplotlib.

**ðŸ§ Interpreting a box plot**

1. The box in a box plot represents the middle 50% of the dataset, also known as the interquartile range (IQR).

2. The line inside the box represents the median of the dataset.

3. The whiskers extending from the box represent the minimum and maximum values that are not outliers.

4. Any points outside of the whiskers are considered outliers and are plotted as individual points.

**ðŸ‘ Advantages of using box plots:**

1. They provide a clear and concise way to visualize the distribution of data, including the presence of outliers.

2. They are useful for comparing distributions between different groups or datasets.

3. They can be easily created using Python and Matplotlib.

**ðŸš« Limitations of using box plots:**

1. They do not show the actual data points, so it can be difficult to see the exact shape of the distribution.

2. They can be misleading if the dataset is not symmetrically distributed or if there are many outliers.

3. They are less effective when the dataset is small.

Overall, box plots are a useful tool for visualizing the distribution of data and identifying potential outliers.

## **Activity 2**

Create a ðŸ“Š bar plot of the average ðŸ’² tip amount for each day of the week.

ðŸ´ðŸ“ŠðŸ“ˆ For this activity, we will be using the tips dataset in Seaborn, which contains information about the amount of tip that customers gave at a restaurant, as well as other variables such as the total bill, day of the week, time of day, etc.

Steps:

ðŸ”¹ Load the tips dataset from Seaborn and group the data by the day variable.

ðŸ”¹ Calculate the mean tip amount for each day using the mean() function.

ðŸ”¹ Create a bar plot of the mean tip amount for each day using the barplot()
function in Seaborn.

### **Solution** ðŸ‘‡

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Define the tips_data variable
tips_data = [
    (16.99, 1.01, 'Female','Mon', 'Dinner', 2),
    (10.34, 1.66, 'Male','Mon', 'Dinner', 3),
    (21.01, 3.5, 'Male', 'Tue', 'Dinner', 3),
    (23.68, 3.31, 'Male', 'Tue', 'Dinner', 2),
    (24.59, 3.61, 'Female','Tue', 'Dinner', 4),
    (25.29, 4.71, 'Male', 'Wed', 'Dinner', 4),
    (8.77, 2.0, 'Male','Thu', 'Dinner', 2),
    (26.88, 3.12, 'Male','Thu', 'Dinner', 4),
    (15.04, 1.96, 'Male','Sun', 'Dinner', 2),
    (14.78, 3.23, 'Male','Sun', 'Dinner', 2)
]

# Convert tips_data to a Pandas DataFrame
tips_df = pd.DataFrame(tips_data, columns=['total_bill', 'tip', 'sex', 'day', 'time', 'size'])

# Group the data by the day variable and calculate the mean tip amount
mean_tip_by_day = tips_df.groupby('day')['tip'].mean()

# Create a bar plot of the mean tip amount for each day
sns.barplot(x=mean_tip_by_day.index, y=mean_tip_by_day.values)

# Customize the x-axis and y-axis labels
plt.xlabel('Day of the Week')
plt.ylabel('Mean Tip Amount ($)')

# Add a title to the plot
plt.title('Mean Tip Amount by Day of the Week')

# Display the plot
plt.show()
````

**Explanation:**
